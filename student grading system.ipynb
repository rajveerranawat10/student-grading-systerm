```python
# F1
import csv
import string 
import random
import pandas as pd
import os
city_codes = []
school_codes = []

# first five school codes and city codes
for i in range(5):
    city = ''.join(random.choice(string.ascii_uppercase) for i in range(3))
    school = ''.join(random.choice(string.digits) for i in range(3))
    city_codes.append(city)
    school_codes.append(school)

f1 = {'city_codes' : city_codes, 'school_codes' : school_codes}
cityAndSchoolCodes = pd.DataFrame(f1)
print (cityAndSchoolCodes)
if not os.path.exists("f1.csv"):
    cityAndSchoolCodes.to_csv("f1.csv",index=False)
    print("CSV file created!!!")
else:
    print("File already exists!!!")
```

```python
# F2
year = 20
examNo = []
stu_name = []
stu_add = []
phoneNo = []
size = 602 # can also take as input

cityAndExamNo, schoolAndExamNo = {}, {}
for i in range(1,size):
    city = random.choice(city_codes)
    school = random.choice(school_codes)

    if(i>=1 and i<=9):
        exm_no='20' + city + school + '00' + str(i)
    elif(i>=10 and i<=99):
        exm_no='20' + city + school + '0' + str(i)
    else:
        exm_no='20' + city + school + str(i)
            
    name = ''.join(random.choice(string.ascii_letters) for i in range(5))
    add = ''.join(random.choice(string.ascii_lowercase) for i in range(10))
    phone = ''.join(random.choice(string.digits) for i in range(10))
    
    if city in cityAndExamNo.keys():
        cityAndExamNo[city] += [str(exm_no)]
    else:
        cityAndExamNo[city] = [str(exm_no)]
    
    if school in schoolAndExamNo.keys():
        schoolAndExamNo[school] += [str(exm_no)]
    else:
        schoolAndExamNo[school] = [str(exm_no)]
        
    examNo.append(exm_no)
    stu_name.append(name)
    stu_add.append(add)
    phoneNo.append(phone)
    
f2 = {'examNo' : examNo, 'stu_name' : stu_name, 'stu_add' : stu_add, 'phoneNo' : phoneNo}
f2Data = pd.DataFrame(f2)
if not os.path.exists("f2.csv"):
    f2Data.to_csv("f2.csv", index=False)
    print("CSV file created!!!")
else:
    print("file already exists!!!")
```

```python
# F3
import numpy as np

sub1, sub2, sub3, sub4, sub5, sub6 = [], [], [], [], [], []
result = []
grad = []
for i in range(1,size):
    s1=random.randint(1,100)
    s2=random.randint(1,100)
    s3=random.randint(1,100)
    s4=random.randint(1,100)
    s5=random.randint(1,100)
    s6=random.randint(1,100)
    res=(s1+s2+s3+s4+s5+s6)/6
    if(res>=90 and res<=100):
        grade='A+'
    elif(res>=80 and res<90):
        grade='A'
    elif(res>=70 and res<80):
        grade='B+'
    elif(res>=60 and res<70):
        grade='B'
    elif(res>=50 and res<60):
        grade='C+'
    elif(res>=40 and res<50):
        grade='C'
    else:
        grade='F'
    
    sub1.append(s1)
    sub2.append(s2)
    sub3.append(s3)
    sub4.append(s4)
    sub5.append(s5)
    sub6.append(s6)
    grad.append(grade)
    result.append(res)
f3 = {'examNo':examNo, 'sub1':sub1, 'sub2':sub2, 'sub3':sub3, 'sub4':sub4, 'sub5':sub5, 'sub6':sub6, 'result':result, 'grade':grad}
f3Data = pd.DataFrame(f3)

if not os.path.exists('f3.csv'):
    f3Data.to_csv('f3.csv',index=False)
    print("CSV file created!!!")
else:
    print("file already exists!!!")
```

```python
import tkinter as tk                
from tkinter import font  as tkfont 
import pandas as pd
from tkinter import ttk

class My_GUI(tk.Tk):

    def __init__(self, *args, **kwargs):
        tk.Tk.__init__(self, *args, **kwargs)
        
        # Make the Font 'Helvetica' and size = 18 weight = bold and italic
        self.title_font = tkfont.Font(family='Helvetica', size=18, weight="bold", slant="italic")

        #tk.Frame makes a frame, Frame helps to visualize the content.
        # tk.pack() is used to load the Frame/Button/Text/etc on the main screen of tkinter
        container = tk.Frame(self)
        container.pack(side="top", fill="both", expand=True) 
        
        # .grid_rowconfigure and .grid_columnconfigure is use to just configure the grid into the screen
        # weight attribute in .grid_columnconfigure does this, Also same thing appear for row_configure but instead the row will grow
        #     A relative weight used to distribute additional space between columns. 
        #     A column with the weight 2 will grow twice as fast as a column with weight 1. 
        #     The default is 0, which means that the column will not grow at all.
        container.grid_rowconfigure(0, weight=1)
        container.grid_columnconfigure(0, weight=1)

        self.frames = {}
        # This is used to set the pages in the tkinter screen
        for F in (StartPage, Page_2):
            page_name = F.__name__
            frame = F(parent=container, controller=self)
            self.frames[page_name] = frame
            frame.grid(row=0, column=0, sticky="nsew")

        self.show_frame("StartPage")
    # The starting page is just the welcome page and can be load using this function
    def show_frame(self, page_name):
        '''Show a frame for the given page name'''
        frame = self.frames[page_name]
        frame.tkraise()

class StartPage(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        
        # tk.Label is used to write Label
        label = tk.Label(self, text="Welcome to PSC Sessional Assignment", font=controller.title_font)
        label.pack(side="top", fill="x", pady=20)
        
        # tk.Button is used to but the button inside our tkinter screen when we pack it.
        # And this function call the lambda function which shifts the controll of tkinter to 
        button1 = tk.Button(self, text="Start",
                            command=lambda: controller.show_frame("Page_2"))
        button1.pack()



class Page_2(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        
        # Displays the Label
        label = tk.Label(self, text="Select the Option to display accordingly", font=controller.title_font)
        label.pack(side="top", fill="x", pady=10)
        
        # Set the Background
        self.config(background="#E0FFFF")
        label.config(background='#9E97CD')
        
        # ------ For the particular Examination number --------
        self.Exam_no_frame = tk.Frame(self)
        self.infoToUser = tk.Text(self, height=1, width=50, bg='#FF5F3E')
        self.infoToUser.pack(side=tk.TOP)
        self.exam_no_entry = tk.Text(self, height=1, width=15, bg='#C8EEF5')
        self.exam_no_entry.pack(side=tk.TOP)

        self.infoToUser.insert(1.0, "Enter the Examination if 3rd option is selected.")
        # tk.DISABLED is used to restrict the user to input something its same as label but rather it uses the tk.Text().pack()
        self.infoToUser.configure(state=tk.DISABLED) 
        self.Exam_no_frame.pack()
        # ------------------------------------------------------
        
        global df
        
        tk.Label(self, text='Select option:').pack()
        
        # Options which we have to used in our tkinter
        options = ['Area Wise', 'School Wise', 'particular student(Examination No. needed!)']
        
        # To use options we use ttk.Combobox to use the Combobox feature in tkinter screen
        self.options = ttk.Combobox(self, values=options, width=30)
        self.options.pack()
        
        # when the button is pressed then show_option function is called!.
        tk.Button(self, text='Show option', command=self.show_option).pack()
        
        # used to display our content while various options is selected with some background
        self.text = tk.Text(self, bg='#D4ECDE')
        self.text.pack()
    
        # It will take you to the first page the welcome page. by calling the function show_frame in My_GUI class
        tk.Button(self, text="Restart",
                  command=lambda: controller.show_frame("StartPage")).pack()

    def show_option(self):
        
        # self.options.get() is used to get the option selected from Combobox
        identifier = self.options.get() # get option
        self.text.delete(1.0, tk.END)   # empty widget to print new text
        data = ""
        
        ls = [False for i in range(601)]
        dfData = np.array(ls)
        
        # if the option area-wise is selected then the dictionary name 'cityAndExamNo' is used in which it store
        # the data of how many Examination no. have the particular city in the form of list
        # so by iterating this list we get our group as well as from this we can fetch whole row from 'allData'(DataFrame)
        if identifier == 'Area Wise':
            self.exam_no_entry.delete(1.0, tk.END)
            # Area wise data made and stored
            for city in cityAndExamNo.keys():
                data += "City Name = " + str(city) + '\n'
                for exam_no in cityAndExamNo[city]:
                    dfData = np.logical_or(dfData, np.array(df['examNo'].str.contains(str(exam_no))))
                data += str(df[dfData]) + '\n'
         
        # Same logic which applied to the area-wise is used here the difference is the dictionary name.
        elif identifier == 'School Wise':
            self.exam_no_entry.delete(1.0, tk.END)
            # School wise data made and stored
            for school in schoolAndExamNo.keys():
                data += "School No. = " + str(school) + '\n'
                for exam_no in schoolAndExamNo[school]:
                    dfData = np.logical_or(dfData, np.array(df['examNo'].str.contains(str(exam_no))))
                data += str(df[dfData]) + '\n\n'
        else:
            # particular student(Examination No. needed!)
            exam_no = str(self.exam_no_entry.get('1.0', tk.END)).strip()
            if exam_no == "": # if the input is empty string then no entry is fetched!.
                data = 'Please Enter Examination No. to get Information!!!'
            else:
                data = df[df['examNo'].str.contains(str(exam_no))]
            
                if data.empty: # if the DataFrame is empty means no data exists in DataFrame so it is invalid number
                    data = 'Invalid Examination No.!!\nEnter the valid Examination No.!!!'
            
        self.text.insert(tk.END, str(data))
            
df = pd.DataFrame(allData) # The DataFrame which is used to store whole data after Natural joining. see above cell for more detail.


if __name__ == "__main__":
    app = My_GUI() # Making the tkinter
    app.mainloop() # Exiting the tkinter when cross(x) is pressed.
```

```python
# 1)
import numpy as np
import matplotlib.pyplot as plt
s1 = allData['sub1']
s2 = allData['sub2']
s3 = allData['sub3']
s4 = allData['sub4']
s5 = allData['sub5']
s6 = allData['sub6']
# AVERAGE, MAXIMUM AND MINIMUM out of the data 
avgSubWise, maxSubWise, minSubWise = [], [], []

avgSubWise += [np.mean(s1), np.mean(s2), np.mean(s3), np.mean(s4), np.mean(s5), np.mean(s6)]

maxSubWise += [np.max(s1), np.max(s2), np.max(s3), np.max(s4), np.max(s5), np.max(s6)]

minSubWise += [np.min(s1), np.min(s2), np.min(s3), np.min(s4), np.min(s5), np.min(s6)]

plt.title("Statistic According to Subject Wise.")
plt.xlabel('Subjects')
plt.ylabel("Marks")
plt.plot(avgSubWise, linestyle='-', color='g', label='Avg-Subject-Wise')
plt.plot(maxSubWise, linestyle='-', color='r', label='Max-Subject-Wise')
plt.plot(minSubWise, linestyle='-', color='b', label='Min-Subjec-Wise')
plt.legend(loc='upper left')
plt.show()
```
